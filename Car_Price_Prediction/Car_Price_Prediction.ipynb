### Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

### Data Collection and Processing

```python
# loading the CSV files into pandas DataFrame
car_dataset = pd.read_csv("car data.csv")
```

```python
# print first 5 rows
car_dataset.head()
```

```python
# checking the number of rows and colums
car_dataset.shape
```

```python
# Getting information about the dataset
car_dataset.info()
```

```python
# Checkin the number of missing values
car_dataset.isnull().sum()
```

```python
# Checking the distributuin of categorical data
print(car_dataset.Fuel_Type.value_counts())
print()
print(car_dataset.Seller_Type.value_counts())
print()
print(car_dataset.Transmission.value_counts())
```

### Encoding the categorical data

```python
# Encoding the "Fuel_Type" column
car_dataset.replace({"Fuel_Type": {"Petrol":0, "Diesel":1, "CNG":2}}, inplace = True)

# Encoding the "Seller_Type" column
car_dataset.replace({"Seller_Type":{"Dealer":0, "Individual":1}}, inplace = True)

# Encoding the "Transmission" column
car_dataset.replace({"Transmission":{"Manual":0, "Automatic":1}}, inplace = True)
```

```python
car_dataset.head()
```

### Spliting the data and target

```python
X = car_dataset.drop(["Selling_Price", "Car_Name"], axis=1)
Y = car_dataset["Selling_Price"]
```

```python
print(X)
```

```python
print(Y)
```

### Split the training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
```

### Model Training
#### 1. Linear regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train, Y_train)
```

### Model Evaluation

```python
# Prediction on training data
training_prediction = lin_reg_model.predict(X_train)
```

```python
# R Squared Error
error_score = metrics.r2_score(Y_train, training_prediction)
print("R Squared Error:", error_score)
```

### Visualize the actual prices and predicted prices

```python
plt.scatter(Y_train, training_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs Pricted prices")
plt.show()
```

```python
# Test data prediction
test_prediction = lin_reg_model.predict(X_test)

# R Squared Error
test_error_score = metrics.r2_score(Y_test, test_prediction)

print("Test data - R Squared Error:", test_error_score)
```

```python
plt.scatter(Y_test, test_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs Pricted prices")
plt.show()
```

### Model Training
#### 2. Lasso regression

```python
# loading the Lasso regression model
lasso_reg_model = Lasso()
```

```python
lasso_reg_model.fit(X_train, Y_train)
```

### Model Evaluation

```python
# Prediction on training data
training_prediction = lasso_reg_model.predict(X_train)
```

```python
# R Squared Error
error_score = metrics.r2_score(Y_train, training_prediction)
print("R Squared Error:", error_score)
```

### Visualize the actual prices and predicted prices

```python
plt.scatter(Y_train, training_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs Pricted prices")
plt.show()
```

```python
# Test data prediction
test_prediction = lasso_reg_model.predict(X_test)

# R Squared Error
test_error_score = metrics.r2_score(Y_test, test_prediction)

print("Test data - R Squared Error:", test_error_score)
```

```python
plt.scatter(Y_test, test_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted price")
plt.title("Actual prices vs Pricted prices")
plt.show()
```
